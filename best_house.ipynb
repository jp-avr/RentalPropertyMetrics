```python
import pandas as pd
```

```python
house_rent = pd.read_excel('houses_rent.xlsx')
```

```python
# Function to display the title and the first rows of a CSV or Excel file
house_rent.head()
```

```python
# Function to display the last rows of a CSV or Excel file
house_rent.tail()
```

```python
# Renames the columns to the value of my choice
house_rent = house_rent.rename(columns={'parking spaces':'parking_spaces','rent amount':'rent_amount','property tax':'property_tax','fire insurance':'fire_insurance'})
```

```python
house_rent.head()
```

```python
# Function to display the data types of each column
house_rent.dtypes
```

```python
# The ASTYPE function is used to change the data type of a specific column
house_rent = house_rent.astype({'rent_amount':'float','property_tax':'float','fire_insurance':'float','total':'float'})
```

```python
house_rent.dtypes
```

```python
# The info function serves the same purpose as DTYPE but provides more detailed information
house_rent.info()
```

```python
house_rent[['area','rooms','bathroom','parking_spaces','rent_amount','property_tax','fire_insurance']].describe()
# The describe function is used to get the count, mean, standard deviation, min, max, and the 25%, 50%, and 75% quartiles values.
```

```python
house_rent = house_rent[(house_rent['total']<=4000)]
house_rent.head()
```

```python
house_rent.describe()
```

```python
house_rent = house_rent[(house_rent['total']<=4000) & (house_rent['animal']=='accept')]
house_rent.head()
```

```python
house_rent = house_rent[(house_rent['total']<=4000) & (house_rent['animal']=='accept') & ((house_rent['rooms']==2) | (house_rent['rooms']==3))]
house_rent.head()
```

```python
house_rent = house_rent.query("total <= 4000 and animal == 'accept'")
house_rent.head()
```

```python
house_rent = house_rent.query("total <= 4000 and animal == 'accept' and (rooms == 2 or rooms == 3)")
house_rent.head()
```

```python
agg_city_price = house_rent.groupby(['city'])['total'].agg("mean").reset_index()
agg_city_price
```

```python
agg_bath_room_price = house_rent.groupby(['bathroom','rooms'])['total'].agg('mean').reset_index()
agg_bath_room_price
```

```python
# Seaborn to plot graphs
import seaborn as sns
```

```python
ax = sns.barplot(data=agg_bath_room_price, x="rooms", y='total');

ax.set(xlabel= "Total number of rooms", ylabel="Average Prices");
```

```python
ax = sns.barplot(data=agg_bath_room_price, x="bathroom", y='total', color='green');
ax.set(xlabel= "Total number of Bathrooms", ylabel="Average Prices");
```

```python
agg_area = house_rent[['area']]
agg_area = agg_area.query("area <= 100")
```

```python
sns.histplot(data=agg_area, x='area')
```
